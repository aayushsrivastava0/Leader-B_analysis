```python
import pandas as pd
import pandas as pd
import numpy as np
import scipy
from scipy.stats import skew
from scipy.stats import kurtosis
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows',None)
```

```python
yearss = [2010, 2011, 2012, 2013,2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
yearss
```

```python
random=r'C:\Users\Aayush Srivastava\Desktop\NEW DATA ANALYSIS\Mean_dataset\Mean_webinar_score.csv'
random_data= pd.read_csv(random)
random_data.shape
```

```python
random_data.studio_name.unique()
```

```python

fig =plt.figure(figsize =(15,15))


for i in yearss:
    DF = random_data[((random_data['year'] == i) &(random_data['studio_name']== 'AI/ML Studio'))]

    
   
    plt.scatter(DF['quarter'], DF['webinar_score'], label = i )
    


plt.legend(fontsize=10)
plt.xlabel('Quarters')
plt.ylabel('webinar_score')

plt.savefig('AIML.png')
```
